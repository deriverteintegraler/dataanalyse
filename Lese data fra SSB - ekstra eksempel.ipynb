```python
geografiskenhet = []
innflyttingtotalt = []
innflyttingfrautland = []
innflyttinginnenlandsk = []
utflyttingtotalt = []
utflyttingtilutland = []
utflyttinginnenlandsk = []
nettoinnflyttingtotalt = []
nettoinnflyttingfrautland = []
nettoinnflyttinginnenlandsk = []
```

```python
for line in data[11:]:
    print(line)
```

```python
for line in data[11:]:
    print(line)
    line = line.rstrip(";").split(";")
    print(line)
    geografiskenhet.append(line[0])
    innflyttingtotalt.append(int(line[1]))
    innflyttingfrautland.append(int(line[2]))
    innflyttinginnenlandsk.append(int(line[3]))
    utflyttingtotalt.append(int(line[4]))
    utflyttingtilutland.append(int(line[5]))
    utflyttinginnenlandsk.append(int(line[6]))
    nettoinnflyttingtotalt.append(int(line[7]))
    nettoinnflyttingfrautland.append(int(line[8]))
    nettoinnflyttinginnenlandsk.append(int(line[9]))
```

```python
print(geografiskenhet)
print(innflyttingtotalt)
```

```python
import matplotlib.pyplot as plt

plt.figure(num=None, figsize=(16, 12), dpi=80, facecolor='w', edgecolor='k')

x = innflyttingtotalt
y = utflyttingtotalt

plt.scatter(x, y)

plt.xlim(0, 1.1*max(max(x),max(y)))
plt.ylim(0, 1.1*max(max(x),max(y)))
    
plt.plot([0, max(max(x),max(y))], [0, max(max(x),max(y))])

plt.xlabel('Innflytting totalt')
plt.ylabel('Utflytting totalt')

#plt.legend(loc='best')
plt.grid()
plt.show()
```
